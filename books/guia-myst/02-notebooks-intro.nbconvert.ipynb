## Gráficos e Visualizações

O MyST suporta nativamente a renderização de gráficos gerados por bibliotecas como Matplotlib, Altair, Plotly e Bokeh. Abaixo, um exemplo simples usando Matplotlib.

```python
# Gerando dados
x = np.linspace(0, 10, 100)
y = np.sin(x) * np.exp(-0.1 * x)

# Criando o gráfico
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="Oscilação Amortecida", color="purple", linewidth=2)
plt.title("Exemplo de Gráfico Matplotlib")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
