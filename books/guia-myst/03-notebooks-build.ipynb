## Gráficos Estáticos

Gráficos gerados pelo Matplotlib são capturados e incluídos no documento final.

```python
# | label: fig-build-plot

x = np.linspace(0, 10, 100)
y = np.sin(x) * np.exp(-0.1 * x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label="Amortecido")
plt.title("Gráfico Gerado no Build (Input Oculto)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Tratamento de Erros

Podemos permitir que células com erro não parem o build usando a tag `raises-exception`.
