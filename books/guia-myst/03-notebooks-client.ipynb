## Configuração

Para habilitar isso, o `myst.yml` deve conter:
```yaml
project:
  jupyter:
    lite: true
```

Isso baixa e configura o kernel Pyodide automaticamente.

## Teste Interativo

Clique no botão de execução (se habilitado) para rodar o código abaixo no seu navegador.

````python
import platform
import sys

print(f"Plataforma: {sys.platform}")
print(f"Versão Python: {platform.python_version()}")

if "emscripten" in sys.platform:
    print("✅ Você está rodando no navegador via WebAssembly!")
else:
    print("ℹ️ Você está rodando em um ambiente Python nativo.")
````

## Manipulação de Dados no Browser

Bibliotecas como Pandas e NumPy funcionam no Pyodide.

````python
import numpy as np
import pandas as pd

df = pd.DataFrame({"Valor": np.random.randn(5), "Categoria": ["A", "B", "A", "B", "C"]})

print("DataFrame criado no cliente:")
df
````
